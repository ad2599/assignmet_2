# **1. Discuss string slicing and provide examples.**
**answer** - When I work with strings in Python, I often need to extract specific parts of them. String slicing allows me to do this efficiently. It uses the syntax string[start:end:step], where start is the index where slicing begins, end is where it stops (not included), and step determines the increment.

For example:
my_string = "Daemon Targaryen"

Slice from index 0 to 6 (excluding 6)

print(my_string[0:6])  # Output: Daemon

Slice from index 7 to the end

print(my_string[7:])  # Output: Targaryen

Reverse the string

print(my_string[::-1])  # Output: neryagraT nomeaD


# **2. Explain the key features of lists in Python**
**answer** - I use lists when I need an ordered, mutable collection of items. They can hold elements of different types and allow for dynamic resizing.

**Key features:**

Ordered: Elements maintain their order of insertion.

Mutable: I can add, remove, or modify elements.

Heterogeneous: They can contain different data types.

Support for slicing: Just like strings, I can slice lists.

Example:

my_list = [1, 'Daemon', 3.5, True]


# **3. Describe how to access, modify, and delete elements in a list with examples.**
**Answer-** Accessing, modifying, and deleting elements in a list is something I do often.

Access: Use indices to retrieve elements.

my_list = [1, 2, 3, 4]

print(my_list[0])  # Output: 1


Modify: Change elements by assigning new values.

my_list[1] = 99

print(my_list)  # Output: [1, 99, 3, 4]

Delete: Use del, remove(), or pop().

del my_list[0]

my_list.remove(3)

print(my_list)  # Output: [99, 4]

my_tuple = (1, 2, 3)

my_list = [1, 2, 3]


# **5. Describe the key features of sets and provide examples of their use**
**answer-** When I need a collection of unique items, I use sets. They are unordered and unindexed.

Key features:

Unique elements: Duplicates are automatically removed.

Unordered: No guaranteed order.


Set operations: I can perform union, intersection, and difference.


Example:

my_set = {1, 2, 2, 3}

print(my_set)  # Output: {1, 2, 3}

**Intersection**
another_set = {2, 3, 4}

print(my_set & another_set)  # Output: {2, 3}

**answer**
Tuples: I use them for fixed collections of data, such as function arguments or geographic coordinates.

Sets: I use them for membership testing, removing duplicates, and performing

coordinates = (52.516, 13.377)

# **7. Describe how to add, modify, and delete items in a dictionary with examples**
**answer**
I rely on dictionaries to store key-value pairs.

Add: I can add new key-value pairs directly.

my_dict = {'name': 'Daemon'}
my_dict['house'] = 'Targaryen'


Modify: Update values using keys.

my_dict['name'] = 'Rhaenyra'


Delete: Use del or pop().

del my_dict['house']
my_dict.pop('name')



# **8. Discuss the importance of dictionary keys being immutable and provide an example**

Keys in dictionaries must be immutable (e.g., strings, numbers, or tuples) because Python uses hash functions to locate keys efficiently. If keys were mutable, their hash value could change, breaking the dictionary's structure.

Example:

Valid key
my_dict = {(1, 2): "Coordinates"}


Invalid key

my_dict[[1, 2]] = "Coordinates"  # Raises TypeError


This immutability ensures consistent performance and integrity of the dictionary.
